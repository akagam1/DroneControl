```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from MDParser import AnalysisParser
import matplotlib.cm as cm
import numpy as np
```

```python
def plotLines(analysisData, variable):

    sns.set_style('darkgrid')
    mus = list(sorted(analysisData.keys()))
    sigmas = list(sorted(analysisData[mus[0]].keys()))

    values = {}

    plt.figure(figsize=(10, 10))

    for mu in mus:
        values[mu] = []
        for sigma in sigmas:
            values[mu].append(analysisData[mu][sigma]['None'][variable])

        plt.plot(sigmas, values[mu])
    
    plt.xlabel("σ", fontsize=12)
    plt.xticks(sigmas, rotation=90)
    plt.ylabel(variable, fontsize=12)
    plt.legend([f"μ={m}" for m in mus])

```

```python
def plotHeatmap(analysisData, variable):

    sns.set_style('white')

    mus = list(analysisData.keys())
    sigmas = list(analysisData[mus[0]].keys())

    Z = []

    for x in sorted(mus):
        Z.append([])
        for y in sorted(sigmas):
            Z[-1].append(analysisData[x][y]['None'][variable])

    Z = np.array(Z)

    plt.figure(figsize=(10, 10))
    plt.matshow(Z, cmap=cm.coolwarm, fignum=1)
    plt.xticks(range(len(sigmas)), sigmas, rotation=90)
    plt.xlabel("σ")
    plt.yticks(range(len(mus)), mus)
    plt.ylabel("μ")
    plt.colorbar()
```

```python
parser = AnalysisParser()
```

### Model 7

```python
analysisDataModel1 = parser.parseAnalysisData('/home/arjune/Documents/DroneControl/SBAgent/models/base/model6/best_model.zip_eval.md_unbiased_biased.md')
plotHeatmap(analysisDataModel1, 'Success Rate')
plt.xticks(np.arange(0,31,5))
plt.yticks(np.arange(0,31,5))
plt.show()

plotHeatmap(analysisDataModel1, 'Mean Reward')
plt.show()
```
